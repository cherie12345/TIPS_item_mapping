```python
df_2.drop(['group_brand'], axis=True, inplace=True)
```

```python
df = pd.concat([df_1, df_2, df_3], axis =0)
df
```

```python
df = df.drop(['id'], axis=1)
df
```

```python
data_all = df.copy()
```

```python
data_all.drop_duplicates(subset='name', inplace=True)
```

```python
data_all
```

```python
def main():
    precisions = []
    recalls = []
    f1_scores = []
    accuracy = []

    for i in range(len(data_all)):
        print("  ")
        print("## Unique Item ", i)

        data = {'brand1': [data_all['brand'].iloc[i]] * len(data_all),
                'item1': [data_all['name'].iloc[i]] * len(data_all),
                'brand2': data_all['brand'],
                'item2': data_all['name']}
        test_df = pd.DataFrame(data)

        # Apply custom function and evaluate
        result = test_df.apply(lambda row: check_same_item(row), axis=1)
        test_df['answer'] = result.apply(lambda x: x[0])
        _, df = select_names_with_same_group(data_all, data_all['name'].iloc[i])
        act = df['actual'].to_list()
        test_df['actual'] = act
        test_df['result_detail'] = result.apply(lambda x: x[1])

        # Compute metrics
        p, r, f1, a = compute_metrics(test_df['answer'].to_list(), act)
        precisions.append(p)
        recalls.append(r)
        f1_scores.append(f1)
        accuracy.append(a)

        # Generate and save CSV
        test_df = test_df.sort_values(by='actual', ascending=False)
        filename = f"../result/0702/keyword/final/all_{i}.csv" # your paht
        test_df.to_csv(filename, index=False, encoding='utf-8-sig')

    print(" ")
    print(" ")
    print("**--------Search within a whole Brand is Done!---------**")

    # Save evaluation results to CSV
    evaluation_df = pd.DataFrame({'item': df['name'],
                                  'precision': precisions,
                                  'recalls': recalls,
                                  'f1_scores': f1_scores,
                                  'accuracy' : accuracy})
    evaluation_df = evaluation_df.sort_values(by='f1_scores', ascending=False)
    filename = "../result/0702/result_revised_0.3.csv"  # your path
    evaluation_df.to_csv(filename, encoding='utf-8-sig')
    
```

```python
main()
```
